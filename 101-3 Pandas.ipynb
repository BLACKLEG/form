Le module pandas a été développé pour apporter à Python les outils nécessaires pour manipuler de gros volumes de données. Ce module a pour objectif de devenir le meilleur outil de manipulation de données : à la fois performant, facile d'utilisation et flexible. Le module existe depuis 2008 mais son développement s'est fortement accéléré avec la publication régulière de nouvelles versions, la version 1.0 étant sortie en Janvier 2020. Pandas est aujourd'hui le module le plus utilisé pour gérer des bases de données sous Python.

Les principales fonctionnalités de pandas sont la récupération des données depuis des fichiers (CSV, tableaux Excel, des pages web, HDF5, etc..) et la manipulation de ces données (grouper, découper, alléger, déplacer,...). Pandas introduit la notion de data frame à Python qui est un concept central en analyse de données moderne.
